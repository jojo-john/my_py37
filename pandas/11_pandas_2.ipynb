```python
#DataFrame对象
import pandas as pd
import numpy as np
df_dict = {
    'name':['James','Curry','Iversion'],
    'age':['18','19','20'],
    'national':['us','china','us']
}
#DataFrame对象 会将key作为 每列的列名 
df = pd.DataFrame(data=df_dict,index=['0','1','2'])
print(df)
#获取行和列
print('---获取行和列----')
print(df.shape)
#获取列 的值
print('--获取列的值--')
print(df['name'])
##获取多个列的值
print('--获取多列的值--')
print(df[['name','age']])
#获取行的值 一定要使用切片
print('--获取行的值,要使用切片---')
print(df[0:1])
#获取行索引
print('--获取行索引--')
print(df.index.tolist())
#获取列索引
print('--获取列索引--')
print(df.columns.tolist())
#获取数据类型
print('--获取数据类型--')
print(df.dtypes)
#获取数据维度
print('--获取数据维度--')
print(df.ndim)
#values属性也会以二维ndArray的形式返回DataFrame的数据
print('--values属性---')
print(df.values)
#展示df的概览
print('--df的概览---')
print(df.info())
#head tail方法和Series的一样
'''
df.loc   通过标签索引行数据
df.iloc  通过位置获取行数据
'''
#获取某一行某一列数据
print('--获取某一行某一列的数据--')
print(df.loc['0','name'])
#获取某一行的所有列：
print('--获取某一行所有列--')
print(df.loc['0',:])
#获取某一行多列
print('--获取某一行 间隔多列--')
print(df.loc['0',['name','national']])
#获取间隔 多行 多列
print('--获取间隔 多行 多列--')
print(df.loc[['0','2'],['name','national']])
#获取连续行和间隔的多列
print('---获取连续的行 和 间隔的多列--')
print(df.loc['0':'2'],['name','national'])


##获取一行
print('--位置索引获取1行---')
print(df.iloc[1])
##获取连续多行
print('--获取连续的多行---')
print(df.iloc[0:2])
##获取间隔的多行
print('--获取间隔的多行---')
print(df.iloc[[0,2],:])
##获取某个值
print('--获取某个值---')
print(df.iloc[0,1])
#dataFrame中的排序方法
print('--dataFrame中的排序--')
print(df.sort_values(by='age',ascending=False))
```

```python
#案列赏析
import pandas as pd
import numpy as np
data= pd.read_csv('./BJ_tianqi.csv')
#print(data)  #将数据的第一行作为列索引 
#print(data.shape) #查看数据形状

#获取某一列有多种方法
#data['日期']
#data.loc[:,'日期']
#改变行的索引---以后可能会用日期作为索引(遇到日期 很可能要将日期作为索引值)---可能辉耀统计一个月 或者一个季度，这样取数方便

#改变数据索引 set_index有一个返回值，如果inplace=true不加  那么要其结果赋值给一个变量，这个变量的索引值才是设置好的值；加上之后直接生效
data.set_index(data['日期'],inplace=True)
data.head()
#发现这个数据 的索引开头有个‘日期’ 实际上这是索引的名称
data.index.name=None
data.head()

##求全年最高温度的平均值
###将最高温度 设置成数字 data['最高温度']--相当于一个Series对象,用字符串str里面的一个方法 replace
data.loc[:,'最高温度']= data['最高温度'].str.replace('℃','').astype('int32')
data.loc[:,'最低温度']=data['最低温度'].str.replace('℃','').astype('int32')

###看全年最高温度
result = data.sort_values(by='最高温度',ascending=False)
#获取最高温度的所有信息
print(result.iloc[0,:])
#获取多行 多列数据
print(result.loc[['2019/1/3','2019/1/4','2019/1/5'],['最高温度','最低温度']])

#最高温度大于32
data.loc[data['最高温度']>32]

#最高温度小于30，最低温度大于20
data.loc[(data['最低温度']>20) & (data['最高温度']<30)]

#查询9月份的数据
index_list = data.index.tolist()
result = [i.startswith('2019/9') for i in index_list]
data[result] #或者 data.loc[result,:]
#[i*2 for i in index_list if i>2] # for 后面只能写if判断，不能写 else
#[True if i.startswith('2019/9') else False for i in index_list] 列表生成式中 如何写 if else

##添加一列数据---温差
data.loc[:,'温差'] = data['最高温度']-data['最低温度']
#温差>10 大温差 ；《=10 正常温差
data.loc[:,'温差类型'] = [ '大温差' if value>=10 else '正常温差' for value in data['温差']]
#查看全年有多少天正常温差----value_counts()---进行分类的统计结果
account_wencha = data.loc[:,'温差类型'].value_counts()
print(account_wencha)
print(data.loc[:,'风向'].value_counts())
#data.head()


#排序 
print(data.sort_values(by='温差',ascending=False))
##多值排序 以列表形式
print(data.sort_values(by=['最高温度','温差'],ascending=[False,True]))

```

```python
#空值处理  有2种方式：dropna 删除空值   --空值不影响分析的时候 不用删除；影响分析的时候  可以删除，
#                                    但数据样本比较少的时候，尽量不要去删，可以采用填充的方式
#                  fillna 填充空值
#                  isnull 判断空值
import pandas as pd

#header=None:避免在读取的时候 将第一行作为列索引 header=2 将第2行作为列索引 （索引从0开始）
##csv和excel的区别在于: csv只能创建1个sheet,如果要指定读取哪个sheet 就用sheetname = 'sheet2'
#df = pd.read_excel('./student_excel.xlsx',header=2)
#skiprows->跳过n行开始读取，usecols->从 第几列读到第几列（左闭右开 ）

#df = pd.read_excel('./student_excel.xlsx',skiprows=2,usecols="B:E")
#删除列--全是空值
df = pd.read_excel('./student_excel.xlsx',header=2)
df
print(df)
#axis表示 是行还是列
#how的值有3个 ：any:只要有空值就删掉； all:全是空值就删掉
#删除全是空值的列
print('---删除全是空值的列---')
df.dropna(axis=1,how='all',inplace=True)
print(df)
#删除全是空值的行
print('---删除全是空值的行---')
df.dropna(axis=0,how='all',inplace=True)
df

#填充方式
#df.fillna({'分数':0},inplace=True)
#只填充姓名这一列  method:填充方式 有2个值： ffill(使用前一个值填充)，bfill（使用后一个值填充）
df.loc[:,'姓名'] = df['姓名'].fillna(method='ffill',axis=0)
df
```

```python
#数据的合并 join merge concat
##join 是基于index的连接    dataframe的列
df3=pd.DataFrame({'Red':[1,3,5],'Green':[5,0,3]},index=list('abc'))
print(df3)
df4=pd.DataFrame({'Blue':[1,9,8],'Yellow':[6,6,7]},index=list('cde'))
print(df4)
#join合并  how有这么几个值 left right  outer（把所有的进行合并，相同的就并在一起） inner（只找行索引相同的）
print('--left连接--')
print(df3.join(df4,how='left'))
print('--outer连接--')
print(df3.join(df4,how='outer'))
print('--inner连接--')
print(df3.join(df4,how='inner'))

##起别名 针对两个dataFrame里面有重名的
print('--起别名---')
df3=pd.DataFrame({'Red':[1,3,5],'Green':[5,0,3]},index=list('abc'))
df4=pd.DataFrame({'Blue':[1,9,8],'Green':[6,6,7]},index=list('cde'))
#print(df3.join(df4,how='left')) 这里会有一个suffix的报错
print(df3.join(df4,how='left',lsuffix='l_green',rsuffix='r_green'))
```

```python
# merge 基于column链接
# merge 实际相当于数据库里的join
df1=pd.DataFrame({'名字':list('ABCDE'),'性别':['男','女','男','男','女'],'职称':['副教授','讲师','助教','教授','助教']},index=range(1001,1006))
print(df1)

df2=pd.DataFrame({'名字':list('ABDAX'),'课程':['C++','计算机导论','汇编','数据结构','马克思原理'],'职称':['副教授','讲师','教授','副教授','讲师']},index=[1001,1002,1004,1001,3001])
print(df2)

#没有指定具体哪一列，它会以左侧为基准
print(pd.merge(df1,df2,how='left'))

#指定列 进行合并  默认采用inner的形式
print('--指定单列合并--')
pd.merge(df1,df2,on='名字')
#指定多列进行合并
pd.merge(df1,df2,on=['职称','名字'])
```

```python
# concat 合并相同的的Excel时或者是给dataframe添加行或者列的时候使用
df1 = pd.DataFrame(np.arange(6).reshape(3,2),columns=['four','five'])
df2 = pd.DataFrame(np.arange(6).reshape(2,3),columns=['one','two','three'])
print(df1)
print(df2)
print(pd.concat([df1,df2],sort=True))
#指定axi=0 纵向=按行合并
print('--纵向合并-按行合并--')
print(pd.concat([df1,df2],axis=0,sort=True,ignore_index=True))

print('--横向合并-按列合并--')
print(pd.concat([df1,df2],axis=1,sort=True,ignore_index=True))
```
