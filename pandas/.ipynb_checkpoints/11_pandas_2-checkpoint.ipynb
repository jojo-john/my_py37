```python
#DataFrame对象
import pandas as pd
import numpy as np
df_dict = {
    'name':['James','Curry','Iversion'],
    'age':['18','19','20'],
    'national':['us','china','us']
}
#DataFrame对象 会将key作为 每列的列名 
df = pd.DataFrame(data=df_dict,index=['0','1','2'])
print(df)
#获取行和列
print('---获取行和列----')
print(df.shape)
#获取列 的值
print('--获取列的值--')
print(df['name'])
##获取多个列的值
print('--获取多列的值--')
print(df[['name','age']])
#获取行的值 一定要使用切片
print('--获取行的值,要使用切片---')
print(df[0:1])
#获取行索引
print('--获取行索引--')
print(df.index.tolist())
#获取列索引
print('--获取列索引--')
print(df.columns.tolist())
#获取数据类型
print('--获取数据类型--')
print(df.dtypes)
#获取数据维度
print('--获取数据维度--')
print(df.ndim)
#values属性也会以二维ndArray的形式返回DataFrame的数据
print('--values属性---')
print(df.values)
#展示df的概览
print('--df的概览---')
print(df.info())
#head tail方法和Series的一样
'''
df.loc   通过标签索引行数据
df.iloc  通过位置获取行数据
'''
#获取某一行某一列数据
print('--获取某一行某一列的数据--')
print(df.loc['0','name'])
#获取某一行的所有列：
print('--获取某一行所有列--')
print(df.loc['0',:])
```
